```python
crimes_counts=np.array([50, 80, 30, 70, 90, 40, 60])
print('crime counts:', crimes_counts)
#Basics
mean_counts = np.mean(crimes_counts)
median_counts = np.median(crimes_counts)   
std_dev_counts = np.std(crimes_counts)
var_counts = np.var(crimes_counts)
print('Mean:', mean_counts)
print('Median:', median_counts) 
print('Standard Deviation:', std_dev_counts)
print('Variance:', var_counts)

#Matrix
crimes_counts_per_region=np.array([[50, 80, 30],
                                   [70, 90, 40],
                                   [60, 20, 10]])
print('Mean crime per mounth across regions:', np.mean(crimes_counts_per_region, axis=0))
print('Mean crime per region across mounths:', np.mean(crimes_counts_per_region, axis=1))
print('Overall mean crime:', np.mean(crimes_counts_per_region))
```

# Data Cleaning

```python
print('missing values in each column:')
count=df.isna().sum()
count=count[count>0]
value={
    'count':count,
    'percentage':count/len(df)*100
}
value=pd.DataFrame(value).reset_index().rename(columns={'index':'column_name'})
display(value)
```

```python
duplicates=df.duplicated().sum()
print(f'number of duplicate rows in the dataframe: {duplicates}')
# Let s suppose that we have duplciate rows and we want to remove them
df=df.drop_duplicates()
print('number of duplicate rows in the dataframe after removing duplicates:', df.duplicated().sum())
```

```python
# Data Cleaning - Handling Missing Values
df=df.dropna(subset=['Vict Sex','Vict Descent'])
#Missing values of Weapon Desc is 60% !!!!!!!!! This column is unreliable

df=df.drop('Weapon Desc', axis=1)
print('Total number of missing values is ',df.isna().sum().sum())
```

```python
# Converting
print(f"Date Rptd type: {df['Date Rptd'].dtype}")
print(f"DATE OCC type: {df['DATE OCC'].dtype}")
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'])
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'])
print('After conversion:')
print(f"Date Rptd type: {df['Date Rptd'].dtype}")
print(f"DATE OCC type: {df['DATE OCC'].dtype}")

df['Hour']=(df['TIME OCC'] // 100).astype('int')
```

```python
# Cleaning Categorical Data
print("Unique areas (before cleaning):")
print(df['AREA NAME'].unique()[:10])
print()

df['AREA NAME'] = df['AREA NAME'].str.strip().str.title()
print("Unique areas (after cleaning):") 
print(df['AREA NAME'].unique()[:10])
```

Which hour has the highest frequency of crimes?

Which area has the largest frequency of night crimes (crimes committed between 10pm and 3:59am)?

Identify the number of crimes committed against victims of different age groups. Save as a pandas Series called victim_ages, with age group labels "0-17", "18-25", "26-34", "35-44", "45-54", "55-64", and "65+" as the index and the frequency of crimes as the values.

```python
# Bil code
highest_freq=df['Hour'].value_counts().index[0]
print('Peak Hour :',highest_freq)

#or
h=df['Hour'].value_counts().to_dict()
max(h, key=h.get)

#Bil nadhar

sns.barplot(x=h.keys(),y=h.values())
plt.title('Number of crimes by hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
pm=(df['TIME OCC']>=2200) & (df['TIME OCC']<=2359)
am=(df['Hour']>=0) & (df['Hour']<4) 
night_df=df.loc[(pm | am),['AREA NAME']]
freq=night_df['AREA NAME'].value_counts()
print('Area with most night crimes:', freq.index[0])
sns.barplot(x=freq.values, y=freq.index)
plt.xlabel('Number of Night Crimes')
plt.ylabel('Area Name')
plt.title('Night Crimes by Area (10pm-3:59am)')
plt.tight_layout()
plt.show()
```

```python
ages_Holder=["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]
age_bins=[0, 17, 25, 34, 44, 54, 64, np.inf]
df['Victem Age Category']=pd.cut(df['Vict Age'],bins=age_bins,labels=ages_Holder)
victim_ages=df['Victem Age Category'].value_counts()
print('Most age category comitted crime against: ', victim_ages.index[0])

sns.barplot(x=victim_ages.index, y=victim_ages.values, palette='viridis')
plt.xlabel('Victim Age Category')
plt.ylabel('Number of Crimes')
plt.title('Crimes by Victim Age Category')
plt.show()
```
